Encoding categorical data


```python
df.head()
```

```python
df['Embarked'].unique()
```

```python
df.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'Q':0,'S':1,'C':2}}, inplace = True)
```

seperating feature matrices

```python
X = df.drop(['Survived','Name','Ticket','PassengerId'], axis=1)
Y = df['Survived']
```

Splitting data into training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

Model training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression()
lr.fit(X_train,Y_train)

dtc = DecisionTreeClassifier()
dtc.fit(X_train,Y_train)

gnb = GaussianNB()
gnb.fit(X_train,Y_train)

kn = KNeighborsClassifier()
kn.fit(X_train,Y_train)

rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
```

model evaluation

```python
xpredlr = lr.predict(X_train)
xpreddtc = dtc.predict(X_train)
xpredgnb = gnb.predict(X_train)
xpredkn = kn.predict(X_train)
xpredrf = rf.predict(X_train)
```

```python
 from sklearn.metrics import accuracy_score
```

```python
accuracy_lr= accuracy_score(Y_train,xpredlr)
accuracy_dtc = accuracy_score(Y_train,xpreddtc)
accuracy_gnb = accuracy_score(Y_train,xpredgnb)
accuracy_kn = accuracy_score(Y_train,xpredkn)
accuracy_rf = accuracy_score(Y_train,xpredrf)
```

```python
print('Accuracy score of training data: ',accuracy_lr,accuracy_dtc,accuracy_gnb,accuracy_kn,accuracy_rf)
```

```python
X_testpredlr = lr.predict(X_test)
X_testpreddtc = dtc.predict(X_test)
X_testpredgnb = gnb.predict(X_test)
X_testpredkn = kn.predict(X_test)
X_testpredrf = rf.predict(X_test)
```

```python
X_testaccuracy_lr= accuracy_score(Y_test,X_testpredlr)
X_testaccuracy_dtc = accuracy_score(Y_test,X_testpreddtc)
X_testaccuracy_gnb = accuracy_score(Y_test,X_testpredgnb)
X_testaccuracy_kn = accuracy_score(Y_test,X_testpredkn)
X_testaccuracy_rf = accuracy_score(Y_test,X_testpredrf)
```

```python
print('Accuracy score of training data: ',X_testaccuracy_lr,X_testaccuracy_dtc,X_testaccuracy_gnb,X_testaccuracy_kn,X_testaccuracy_rf)
```
